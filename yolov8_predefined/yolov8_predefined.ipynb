```python
from ultralytics import YOLO
import os

# Define the input image path and the desired output directory
input_image_path = r'C:\Desktop\projects\yolov8_predefined\input1.PNG'
output_directory = r'C:\Desktop\projects\yolov8_predefined'
model_path = r'C:\Desktop\projects\yolov8_predefined\yolov8n.pt'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Load the model
model = YOLO(model_path)

# Run the prediction and save the output to the specified directory
results = model.predict(source=input_image_path, save=True, save_dir=output_directory)

# Print the path to the saved image
print(f"Output saved to: {output_directory}")
```
